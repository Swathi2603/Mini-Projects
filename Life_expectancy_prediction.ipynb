```python
df1 = pd.get_dummies(df1, columns = ['Status'])
```

```python
df1.isnull().sum()[np.where(df.isnull().sum() != 0)[0]]
```

```python
df1 = df1.apply(lambda x: x.fillna(x.mean()),axis=0)
```

```python
X = df1.drop(columns = ['Life expectancy '])
y = df1[['Life expectancy ']]
```

```python
X = np.array(X).astype('float32')
y = np.array(y).astype('float32')
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
from sklearn.preprocessing import StandardScaler

scaler_X = StandardScaler()
X_train = scaler_X.fit_transform(X_train)
X_test = scaler_X.transform(X_test)

scaler_y = StandardScaler()
y_train = scaler_y.fit_transform(y_train)
y_test = scaler_y.transform(y_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, accuracy_score

regresssion_model_sklearn = LinearRegression(fit_intercept = True)
regresssion_model_sklearn.fit(X_train, y_train)
```

```python
accuracy = regresssion_model_sklearn.score(X_test, y_test)
accuracy
```

```python
print('Linear Model Coefficient (m): ', regresssion_model_sklearn.coef_)
print('Linear Model Coefficient (b): ', regresssion_model_sklearn.intercept_)
```

```python
y_predict = regresssion_model_sklearn.predict(X_test)
```

```python
plt.plot(y_test, y_predict, "o", color = 'g')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
plt.title('Results')
```

```python
y_predict_orig = scaler_y.inverse_transform(y_predict)
y_test_orig = scaler_y.inverse_transform(y_test)
```

```python
plt.plot(y_test_orig, y_predict_orig, "o", color = 'r')
plt.xlabel('Model Predictions')
plt.ylabel('True Values')
plt.title('Original')
```
